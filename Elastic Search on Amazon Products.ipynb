```python
import json
# Load your JSON data
with open('dataa.json') as f:
    data = json.load(f)
```

```python
# Index each product in Elasticsearch
for product in data:
    es.index(index='amazon_products_index', body=product["Product Data"])
```

```python
# Refresh the index to make the data available for search immediately (optional)
es.indices.refresh(index='amazon_products_index')
```

```python
# Check if the data is successfully indexed
response = es.search(index='amazon_products_index', body={"query": {"match_all": {}}})
print(response)
```

```python
search_term = "Super Mario Bros"
response = es.search(index='amazon_products_index', body={"query": {"match": {"Product Name": search_term}}})
# Extract hits from the response
hits = response.get('hits', {}).get('hits', [])

# Print the formatted hits
formatted_hits = json.dumps(hits, indent=2)
print(formatted_hits)
```
